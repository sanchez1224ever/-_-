```python
df_clear = df.copy()
```

```python
for col in col_list:
    # Первая квантиль
    Q1 = df[col].quantile(0.25)

    # Третья квантиль
    Q3 = df[col].quantile(0.75)

    # Межквантильное расстояние
    IQR = Q3 - Q1

    max = Q3 + 1.5*IQR
    min = Q1 - 1.5*IQR
    
    df_clear[col] = df[col].apply(lambda x: np.nan if (x > max or x < min) else x)
```

## Линейная регрессия

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
from sklearn.linear_model import LinearRegression
```

```python
def create_train_test_data(col_name):
    X = df_norm.drop(columns= ['Модуль упругости при растяжении, ГПа', 'Прочность при растяжении, МПа']).values
    y = df_norm[col_name].values
    return (train_test_split(X, y, test_size = 0.3, random_state = 42))
```

### Модуль упругости при растяжении

```python
data = 'Модуль упругости при растяжении, ГПа'
```

```python
X_train, X_test, y_train, y_test = create_train_test_data(data)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
y_pred_train = regressor.predict(X_train)
train_rmse = round(mean_squared_error(y_train, y_pred_train), 4)
train_rmse
```

```python
round(mean_squared_error(y_test, y_pred), 4)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(y_pred, color='red', label = 'prediction')
plt.plot(y_test, color='green', label = 'test')
plt.legend(loc="upper left")
plt.savefig('Лин_модуль_упр.png', dpi=300, bbox_inches='tight')  
plt.show
```

### Модуль упругости при растяжении

```python
best_params = gd_sr.best_params_
best_score = gd_sr.best_score_
print(best_params)
print(round(best_score, 4))
```

## Нейронные сети

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
def create_train_test_data_nn(col_name):
    X = df_norm.drop(columns= [col_name]).values
    y = df_norm[col_name].values
    return (train_test_split(X, y, test_size = 0.3, random_state = 42))
```

```python
data = 'Соотношение матрица-наполнитель'
X_train, X_test, y_train, y_test = create_train_test_data_nn(data)
```

```python
model = Sequential()
model.add(Dense(64, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dense(64, activation='relu'))
model.add(Dense(1, activation='sigmoid')) # sigmoid, т.к. данные от 0 до 1
```

```python
model.summary()
```

```python
X_train.shape
```

```python
model.compile(optimizer='adam', loss='mse', metrics=['mae'])
```

```python
prediction = model.fit(X_train, 
                    y_train, 
                    epochs=50, 
                    validation_split=0.1, 
                    verbose=2)
```

```python
ep = np.arange(1, 51)
plt.figure()
plt.plot(ep, prediction.history["loss"], label="train_loss")
plt.plot(ep, prediction.history["val_loss"], label="val_loss")
plt.title("Training Loss")
plt.xlabel("Epoch #")
plt.ylabel("Loss")
plt.legend()
```

```python
model.save('model_nn')
```

```python
y_pred = model.predict(X_test).flatten() 
```

```python
round(mean_squared_error(y_test, y_pred), 4)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(y_pred, color='red', label = 'prediction')
plt.plot(y_test, color='green', label = 'test')
plt.legend(loc="upper left")
plt.savefig('Нейронная сеть.png', dpi=300, bbox_inches='tight')
plt.show
```
